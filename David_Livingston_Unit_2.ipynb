```python
## Inititallize and call tools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Data source: kaggle.com Human Reources Analytics data set
dfHR = pd.read_csv('HR_comma_sep.csv')
```

```python
## Defining dataframes for visualizations 

sat, last, hours = dfHR['satisfaction_level'], dfHR['last_evaluation'], dfHR['average_montly_hours']
time, left = dfHR['time_spend_company'], dfHR['left']
```

```python
## Data set desription/information
dfHR.info()
```

```python
## The purpose of the following visualizations is to evaluate job satisfaction
## At a high level will be taking a look at data to see if there are visual indicators to see if there are drivers of employee satisfaction
## Visualization, while helpful as a starting point, needs to be followed by statistical analysis to determine corrleations
```

```python
# Boxplot to identify relationship between titme spent with company and relative satisfaction

df1 = dfHR.groupby('time_spend_company')

df1.boxplot(column='satisfaction_level',layout=(2,10), figsize=(20,10), whis=[5,95])

plt.show()
```

```python
## The above visulization seems to indicate that employee satisfaction might be related to tenure
## In early years, while satisfaction is relatively high, this might be driven by outliers, and satisfaction declines until
## the fifth year.  This would indicate that further anlalysis is warranted to determine what other factors might be driving 
## the disatisfaction in the early years.  Surveys could be conducted to identify determinants of satisfaction or disatisfaction
## in the early years.  Further analyis is requred before drawing full conclusions
```

```python
## Histogram to show the distribution of satisfaction scores amongst employees

plt.hist(sat, normed = True)
plt.title('Satisfaction Distribution')
plt.show()
```

```python
## The distibution of satisfaction scores appears to be bi-modal with a group of  employees dissatisfied and then
## a larger group of employyess with satisfaction scores in a distribution of values between 0.4 and 1.0, skewed slightly
## to the right.  This visualization indicates that there is a set of employees who are somewhat dissatisfied and further
## analysis is required to understand the driver of disatisfaction.  However, the majority of employees do have a satisfaction 
## > 0.6, signifying most employees are satisfied.   Management should understand drivers of satisfaction or dissatisfaction and 
## conduct more analysis to determine appropriate actions 
```

```python
## Sublots to explore satisfaction versus review score

plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.plot(sat, color='red')
plt.ylabel('Satisfaction')
plt.title('Satisfaction Scores by Respondent')

plt.subplot(1, 2, 2)
plt.scatter(x = sat, y = last, color='green')
plt.ylabel('Review Score')
plt.xlabel('Satisfaction')
plt.title('Satisfaction vs Review Score')

plt.tight_layout()
plt.show()
```

```python
## The left graph shows the distribution of satisfaction scores by respondent and the right  graph shows satisfaction score
## versus the review score.  The right graph, overall indicates that there is a relationship between satisfaction and employee
## review score.  Thre are two interesting clusters that are worth noting.  The cluster in the upper left corner shows a high
## satisfaction score with low review score.  This might reflect new employees who have not yet recieved their review. It 
## would be interesting to separate explore employee satisfaction in their first year to determine drivers of satisfaction.
## The other intersetign cluster is the employees with satisfaction scores of approximately 0.375 and 0.5.  This group seems
## to be generally dissatisfied, and further exploration on this group is warranted.
```

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.plot(sat, color='blue')
plt.ylabel('Satisfaction')
plt.title('Satisfaction Scores by Respondent')

plt.subplot(1, 2, 2)
plt.scatter(x = sat, y = hours, color='purple')
plt.ylabel('Hours Worked')
plt.xlabel('Satisfaction')
plt.title('Satisfaction vs Hours Worked')

plt.tight_layout()
plt.show()
```

```python
## The left graph shows the distribution of satisfaction scores by respondent and the right  graph shows satisfaction score
## versus the monthly hours worked.  The right graph, overall indicates that there is a relationship between satisfaction and 
## hours worked.   It is interesting to note that the satisfaction versus hours worked visualization demonstrates a similar
## pattern as satisfaction vs review score. There is a cluster of employees who work high hours and have a low satisfaction score.
## Thsi might indicate a correleation and is worth exploring, however this dissatisfaction cluster only appears to be relevant
## above approximately 275 hours.  Also, it is interseting to see a similar cluster of individuals who work 140-160 hours with 
## mediocre satisfaction, similar to individuals who received mediocre review scores.
```
